# Visualisations

```python
# Order of different sizes
ax=sns.countplot(x='Size' ,data=df)
plt.show()
# checking for data labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Courier Status
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Courier Status',hue= 'Status')
plt.show()
```

```python
# Most bought Category
df['Category'] = df['Category'].astype(str)
c_d = df['Category']
plt.figure(figsize=(10,6))
plt.hist(c_d, bins=15, edgecolor = 'Black')
plt.xticks(rotation=90)
plt.show()
```

T- shirt is the most bought.

```python
# check b2b data
check_B2B = df['B2B'].value_counts()

#create pie chart
plt.pie(check_B2B, labels=check_B2B.index, autopct = '%1.1f%%')
plt.show()
```

Maximum 99.2% of buyers are retailers and 0.8% are wholesalers 

```python
# Available Sizes
x_data = df['Category']  
y_data = df['Size'] 

# Plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Category ')  
plt.ylabel('Size')  
plt.title('Available Size') 
plt.show()
```

```python
# Distribution of Orders by States
top10_state = df['ship-state'].value_counts().head(10)  # for top 10 states
plt.figure(figsize=(10,6))
sns.countplot(data= df[df['ship-state'].isin(top10_state.index)], x = 'ship-state')
plt.xlabel('State')
plt.ylabel('Order_count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

Most of the buyers are from Maharashtra
